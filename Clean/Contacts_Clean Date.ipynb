```python
contacts_df = pd.read_csv('drive/MyDrive/Final_Project_ICH/CSV_file/Contacts.csv', dtype = {'Id': str})
#contacts_df.info()
contacts_df.head()
```

```python
# проверка дубликатов
contacts_df[contacts_df.duplicated()]
```

```python
# Преобразование формата дат
contacts_df["Created Time"] = pd.to_datetime(contacts_df["Created Time"], format="%d.%m.%Y %H:%M", errors='coerce')
contacts_df["Modified Time"] = pd.to_datetime(contacts_df["Modified Time"], format="%d.%m.%Y %H:%M", errors='coerce')
```

```python
contacts_df['Contact Owner Name'].nunique()
```

```python
# Получение уникальных значений в столбце 'Contact Owner Name'
unique_values = contacts_df['Contact Owner Name'].unique()
unique_values
```

```python
# Замена FALSE на имя ответственного
contacts_df['Contact Owner Name'] = contacts_df['Contact Owner Name'].replace({'FALSE': 'Charlie Davis'})
```
